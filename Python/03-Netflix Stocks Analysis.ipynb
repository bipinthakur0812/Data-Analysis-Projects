# NETFLIX STOCKS ANALYSIS

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
import plotly.graph_objects as go

from datetime import datetime
```

```python
df = pd.read_csv("03-Netflix Stocks Data.csv")
df.head()
```

## Volume of Stock Traded

```python
sns.set(rc={'figure.figsize' : (10,5)})
```

```python
df['Date'] = pd.to_datetime(df['Date'])
df = df.set_index('Date')
df.head()
```

```python
sns.lineplot(x=df.index , y=df['Volume'], label = 'Volume')
plt.title("Volume of Stocks v/s Time")
plt.show()
```

## High, Open and Close of Stocks

```python
df.head()
```

```python
df.plot( y = ['Open','High','Close'], title = " Open, High and Closing of Stocks")
plt.show()
```

## Monthly, Yearly and DayWise Analysis of Volume of Stocks

```python
fig, (ax1,ax2,ax3) = plt.subplots(3, figsize = (15,10))
df.groupby(df.index.day).mean().plot(
                                        y = 'Volume', ax = ax1, xlabel = "Day" , color = 'red')
df.groupby(df.index.month).mean().plot(
                                        y = 'Volume', ax = ax2, xlabel = "Month" , color = 'green')
df.groupby(df.index.year).mean().plot(
                                        y = 'Volume', ax = ax3, xlabel = "Year", color = 'blue' )
plt.show()
```
